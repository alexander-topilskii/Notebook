```python
# prompt: display  x+5y=0 function

import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values
x = np.linspace(-10, 10, 400)

# Calculate the corresponding y values based on the equation x + 5y = 0
y = -x / 5

# Plot the function
plt.plot(x, y)

# Customize the plot (optional)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x + 5y = 0')
plt.grid(True)

# Show the plot
plt.show()
```
